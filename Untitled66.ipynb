```python
df1=DataFrame({'image':yes})
df1['target']=[1]*1500
```

```python
df2=DataFrame({'image':no})
df2['target']=[0]*1500
```

```python
import pandas as pd 
df =pd.concat([df1, df2], ignore_index=True)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
import tensorflow as tf
import pandas as pd
def resize_image(image, target_size=(180, 180)):
    image = tf.convert_to_tensor(image)
    resized_image = tf.image.resize(image, target_size)
    return resized_image
df['resized_image'] = df['image'].apply(resize_image)
```

```python
xtra,xtes,ytra,ytes = train_test_split(df[['resized_image']], df['target'], test_size=0.2, random_state=42)
```

```python
xtra=xtra/255
xtes=xtes/255
```

```python
import numpy as np

# Convert resized images column to a NumPy array
xtra_cleaned = np.array([img.numpy() for img in xtra['resized_image'].values.flatten()])

# Convert labels to a NumPy array of integer type
ytra = np.array(ytra, dtype=np.int64)
```

```python
xtes_cleaned = np.array([img.numpy() for img in xtes['resized_image'].values.flatten()])
ytes = np.array(ytes, dtype=np.int64)
```

```python
from tensorflow import keras
from tensorflow.keras import layers, Sequential

data_aug = Sequential([
    layers.experimental.preprocessing.RandomZoom(0.3),
    layers.experimental.preprocessing.RandomContrast(0.2),
    layers.experimental.preprocessing.RandomRotation(0.3),
])

# Define the model
model = keras.Sequential([
    data_aug,
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(180, 180, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(200, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(xtra_cleaned, ytra, epochs=10)
```

```python
model.evaluate(xtes_cleaned,ytes)
```
